```python
from models import PaddleModel
from PIL import Image

ex_img_fp = '/playpen-storage/levlevi/player-re-id/src/testing/ocr_analysis/sample_tracks_full_length/17706_01-05-2016_3280_milwaukee bucks_20_san antonio spurs_period4.mp4/009229.jpg'
ex_pil_img = Image.open(ex_img_fp)
model = PaddleModel()
```

```python
import numpy as np
from typing import List

def extract_text_with_paddle(image: Image.Image, model: PaddleModel) -> List[str]:
    """
    Returns a [str] containing all words found in a
    provided PIL image.
    """

    if image is None:
        return []
    image = np.array(image)
    
    raw_result = model.model(image)
    return raw_result
```

```python
from glob import glob
from tqdm import tqdm

EXT = '.jpg'

def extract_all_text_from_dir_paddle(dir_path: str, model: PaddleModel) -> List[str]:
    file_paths = glob(dir_path + '/*' + EXT)
    results = []
    for fp in tqdm(file_paths, total=len(file_paths)):
        img = Image.open(fp)
        results.append(extract_text_with_paddle(img, model))
    return results

img_dir = '/playpen-storage/levlevi/player-re-id/src/testing/ocr_analysis/sample_tracks_full_length/55'
results = extract_all_text_from_dir_paddle(img_dir, model)
print(results[0:5])
```
